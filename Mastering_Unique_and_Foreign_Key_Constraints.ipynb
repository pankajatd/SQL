**There are 3 types of Constraints in SQL**

**1. Primary key** --This has two features:
     i) Unique
     ii) NOT NULL

**2. Unique Key** --This has a single feature. NOT NULL contraint is not applied for this key
     i)Unique


**3. Foreign Key** -- This key has two features
     i) Field/Column Nam eto which it is pplied has to be mentioned
     ii) Reference  to the Parent Table and the Primary Key of the table has to be mentioned.

Example:

FOREIGN KEY(SelectedCourses) REFERENCES Course(CourseID)

Note: The Foreign key CourseID is the field to which it is Refrenced to is an Integer, hence the field SelectedCourses..should also be Integer.
