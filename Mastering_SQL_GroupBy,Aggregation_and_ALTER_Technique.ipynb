**Group By**


**GROUP BY ** Command will check for the Unique Values of Enteries in the Specified Column.

Example: If the Data has 3 Enteries for the Column "SourceOf Joining" as
i.LinkedIN

ii.YouTube

iii. Instagram ...

GROUP BY will count the Number of Enteries in each of these  3 Enteries..
ie .. How many enteries in **LinkedIn**,How many in **YouTub**e and how many in **Instagram**.



This command is used to group categories of a particular Column.

**Points to Remember while writing a Query for Group By:**

1. The Column on which you do Group By should be always there after the SELECT statment.
2. After the SELECT statement followed by Column name on which we are grouping there should be a **Aggregation** statement

** 5 Types of SQL Aggregation commands:**

1. COUNT()
2. SUM()
3. MIN()
4. MAX()
5. AVG()

With the GROUP BY Command only the bove 5 Aggregation Commands need to be used.


**Query1: Count the Number of Students who joined the Course Via LinkedIN ,YouTube,COmmunity**

SELECT SourceOfJoining,COUNT(*) as Tot_num_Students_Each_Grp  FROM Learners
GROUP BY SourceOFJoining;

Output:

SourceOfJoining, Total_Studs

'LinkedIn', '3'

'YouTube', '3'

'Instagram', '1'


** Multiple Columns can be used for Group BY. And those multiple column have to be used immedeatly after the SELECT Statement  ,before using the Aggregation statement**


** NOTE --VERY IMPORTANT**

** For Filteration of Data AFTER GROUP BY Clause we can only use **Having** clause and not the **Where** Clause

**Interview Question - What is difference between TRUNCATE and DELETE command in SQL?**

Ans-

**DELETE :-**
Delete is used to remove the specific row from the table using WHERE clause .
It Deletes rowone-by-one.

This is like erasing one at a time in a NoteBook

Example:- DELETE FROM employee WhERE Location="Delhi"


**TRUNCATE**
Truncate is used to remove enite rows of the Table and no need of the WHERE Clause.
This is like throwing away the whole book and getting a new one

TRUNCATE is Fast

Example ; TRUNCATE TABLE employee;
