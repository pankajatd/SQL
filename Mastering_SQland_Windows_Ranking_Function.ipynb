INNER JOIN -- Can join two tables.

Example - Join tables table1 ,table2

The Output of JOIN is the Intersection of two tables..i.e the common entity from the two tables is obtained.

**Example Query: Write a command to find the total number of employees corresponding to each Location and Avg Salary in those Locations
Also display Firstname ,LastName corresponding to each record.**

SELECT FirstName,Lastname ,employee.Location ,Tot_Emp_Per_Loc,Avg_Salary
FROM employee
JOIN
(SELECT Location,COUNT(Location) As Tot_Emp_Per_Loc ,AVG(Salary) As Avg_Salary FROM employee
GROUP BY Location) As temp_tbl
ON employee.Location =temp_tbl.Location;

**3.RANK()**
** We obeserve from the above Output that the employees who have the same Salary have ben given different Priority,which is Not correct. Row_Number() looks at teh EMp_ID and gives the higher Priority ,which comes first.

**DENSE_RANK()**

We obeserve from the above Output that since same Priority is assigned to the People gettingsame Salary ,the Next level is skipped for the number of times it is repeated ,which is not a correct method. To handlethis we we use a method called **Dense_Rank()**.

**QUERY**
/** DENSE_RANK() is USed to handle the issue that is seen with RANK()**/

SELECT FirstName,LastName,Salary,

**INTERVIEW QUESTION-What is the difference between ROW_NUMBER(),RANK() and DENSE_RANK()?**

**Ans: ROW_NUMBER()-will map the values with the Numbers on the basis of Sorted Values.

In Row_Number()-Even though the Salary was same for multiple employees ,it assigned different Priority Level.It assigns the hiher Priority for the Id that comes first.

RANK()- To handle the isssue that occured in Row_Number(),Rank() is used so that it assigns the same Priority of the employees with the same Salary.

But when multiple employees are assigned the same Priority,it skips the consicutive Number, for which the number of times the Present Level is repeated.

DENSE_RANK()- To handle the isssue that occured in RANK(),Dense_Rank() is used ,so that it assigns the same Priority of the employees with the same Salary.when multiple employees are assigned the same Priority,it does not skip the consicutive Number, for which the number of times the Present Level is repeated.**


**NOTE: IF there are no diuplicate records, ROW_NUMBER(),RANK(),DENSE_RANK() will all work the same ie it Returns the SAME ANSWER**
