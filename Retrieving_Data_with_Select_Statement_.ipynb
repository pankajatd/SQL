**Create table with NOT NULL CONSTRAINT**



The column with **NOT NULL** constarint cannot be left empty without entering the data  while inserting the data into the column.

This done to avoid **"Null**" values in a specific column.

Example:

CREATE TABLE employee(
        FirstName varchar(20) NOT NULL,
		    lastName varchar(20) NOT NULL,
        Age int NOT NULL,
        Salary int NOT NULL,
        Location varchar(20) NOT NULL
);

To see the Table description that is created with the constraint:


DESC employee;


we see the table created as below:

FirstName	varchar(20)	NO		NULL
lastName	varchar(20)	NO		NULL
Age	int	NO			NULL
Salary	int	NO		NULL
Location	varchar(20)	NO	NULL

**Error seen when data is not entered for all columns with NOT NULL Constraint**

With the constraint all the column have to be filled while inserting the data into the table columns..

We will see the below error if any of the data is missing in any of the columns:


Example:

output Error: 15:31:54	INSERT INTO employee (lastName,Age,Salary,Location)VALUES("Bhatt",29,50000,"Gurugram")	Error Code: 1364. Field 'FirstName' doesn't have a default value	0.015 sec

This is seen because we donot have the data for the FirstName column



**Correct way to enter the column data for a table with NOT NULL constraint**

**Filtering the Records using WHERE clause**

This command allows to retrieve/dispaly the records from the table that satisfies the conditions.

Example:

select FirstName,LastName from employee WHERE Salary>=50000

** Order By Clause** - This coomand allows us to retrieve the data from the table in the Ascending Order for a particular column.

The Order can be changed to Descending by mentioning the Keyword DESC while retrieving the records .


Example:


SELECT FirstName,LastName,Salary from employee where Salary>=50000 ORDER BY Salary DESC;

Output:

'Alia', 'Bhatt', '50000'
'Ankit', 'Sangwan', '70000'

**Understanding the ORDER OF EXECUTION of the command with Where  and Order By clause**


Example Command:


SELECT FirstName,LastName,Salary from employee where Salary>=50000 ORDER BY Salary;


***Order of Execution: ***

**from ==> where==> Select==>order by**

1. From clause is executed-->It finds from which Table to Fetch the Data
2. Where clause is Executed--> It filters teh data based on the Where Clause.
3. Select command is executed--> It fetches the specified column data.
4. Order by clause is executed--> It displays the fetched records in the specified Order.Either "Ascending" or "Descending". By Default it is "Ascending"



**Extract the Top Two Employees who get the Highest Salary using LIMIT Clause**

The LIMIT clause allows us to specify the top number of data we want to retrieve.


Example:

SELECT FirstName,LastName,Salary from employee where Salary>=50000 ORDER BY Salary DESC LIMIT 2;



***Order of Execution: ***

**from ==> where==> Select==>order by ==> LIMIT**
