**Steps for creating the Database and Inserting .csv file Data into the table**

1. Create a Database.
2. Open the Anaconda prompt: In the Anaconda Prompt provide the path where the Excel file is present.
3. Install the csvkit. The command ti install is "pip install csvkit".
This command will convert the csv file into SQL format.
4. Once csvkit is installed, wneed to write the below command in the command prompt to convert csv file tosql format.

>>csvsql --dialect mysql --snifflimit 10000 Sales_Dataset.csv > Result.sql

A result .sql file is created in the the path where the source .csv file "Sales_Dataset.csv" is present.

5.Open the Result.sql as Notepad..Copy-Paste the contents into MySQL Workbench.

**Output command**

CREATE TABLE `Sales_Dataset` (

6. **Load the data available in Sales_Datset.csv file to the Sales_dataset table**

Below command is used to load the Sales_data.csv file into the table that is created..

**NOTE** : Place the .csv file in the path where MySQL is installed as seen below.
