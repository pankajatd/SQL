**What is the difference between the UPDATE command and the ALTER command?**

A:**ALTER** - This command is used to change the **SCHEMA** of the table

**ALTER** has three types of subcommands that are used along with **ALTER**

i. **ADD** -Add a New Column

Example -

**To add an Additional column in the table employee we use ALTER  command with ADD**

Example-

ALTER TABLE employee ADD Job_Desc varchar(20) ;

ii. **MODIFY**-Modification of the already available column

Example-

Modifying the schema of the employee table . change the length of the column FirstName from varchar(50 ) to varchar(20)

ALTER TABLE employee **MODIFY** FirstName varchar(20) ;

iii. **DROP**-Delete any specific column

Example:

**To delete an existing column in the table employee we use ALTER  command with DROP**

Example:

ALTER TABLE employee DROP Salary int;

  **UPDATE** -This command is used to change the data of any available records in the table.

  Example:

  Update employee set FirstName =" Tiwari" WHERE EmpID=4

**Aggregation Commands** -All the actions are taken on the table in a Database.

1. COUNT()
2. MAX()
3. MIN()
4. SUM()
5. GROUP BY()

Example:

1. **Count command** - Count the number of employees working in a company in the employee table

Example:

SELECT COUNT(*) FROM employee; --This command counts the number of employees in the table.

**Provide the alias name to the column COUNT(*)**

Example:

SELECT COUNT(*) As Total_Num_Employee FROM employee;

Output:

Total_Num_Employee
8



2. **Max command** - Provides  Maximum value for a particular column in the employee table

Example:

SELECT MAX(Salary) FROM employee; --This command returns the Max Salary of employees in the table.

**Provide the alias name to the column Max(Salary) As Max_Salary**

Example:

SELECT MAX(Salary) As Max_Salary FROM employee;

Output:

Max_Salary
'100000'


3. **Min command** - Provides  Minimum value for a particular column in the employee table

Example:

SELECT MIN(Salary) FROM employee; --This command returns the Min Salary of employees in the table.

**Provide the alias name to the column Min(Salary) As Max_Salary**

Example:

SELECT MIN(Salary) As Min_Salary FROM employee;

Output:
Min_Salary
'10000'

4. **Sum command** - Provides  total aggregated value for a particular column in the employee table

Example:

SELECT SUM(Salary) FROM employee; --This command returns the total  SUM  Salary of all the employees in the table.

**Provide the alias name to the column Sum(Salary) As Total_Cost_Company**

Example:

SELECT SUM(Salary) As Total_Cost_Company FROM employee;
